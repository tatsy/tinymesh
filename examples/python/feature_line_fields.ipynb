# Feature line fields

```python
import os
import numpy as np
from matplotlib import cm
```

```python
from tinymesh import Mesh, get_feature_line_field
from viewer import Viewer
```

## Load mesh file

```python
models_dir = "../../data/models"
filename = os.path.join(models_dir, "fandisk.ply")
mesh = Mesh(filename)
mesh.fill_holes()
```

## Compute feature line field

```python
flf = get_feature_line_field(mesh)
```

```python
viewer = Viewer(800, 500)
viewer.lookat(5, 5, -5, 0, 0, 0, 0, 1, 0)
viewer.perspective(45, viewer.aspect, 1, 100)
viewer.mesh_visualization(mesh, colors=flf * 0.5 + 0.5)
```
