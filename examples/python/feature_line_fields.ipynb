# Feature line fields

```python
import os
import numpy as np
from matplotlib import cm
```

```python
from tinymesh import Mesh, get_feature_line_field
from viewer import Viewer
```

## Prepare viewer

```python
viewer = Viewer(800, 500)
viewer.set_identity()
viewer.lookat(7, 5, 7, 0, 0, 0, 0, 1, 0)
viewer.perspective(45, viewer.aspect, 1, 100)
```

## Torus

```python
models_dir = "../../data/models"
filename = os.path.join(models_dir, "torus.ply")
mesh = Mesh(filename)
mesh.fill_holes()
```

```python
flf = get_feature_line_field(mesh, smooth_tensors=True)
viewer.mesh_visualization(mesh, colors=flf * 0.5 + 0.5)
```

## Fandisk

```python
models_dir = "../../data/models"
filename = os.path.join(models_dir, "fandisk.ply")
mesh = Mesh(filename)
mesh.fill_holes()
flf = get_feature_line_field(mesh, smooth_tensors=True)
```

```python
viewer.set_identity()
viewer.lookat(5, 5, 5, 0, -1, 0, 0, 1, 0)
viewer.perspective(45, viewer.aspect, 1, 100)
viewer.mesh_visualization(mesh, colors=flf * 0.5 + 0.5)
```
